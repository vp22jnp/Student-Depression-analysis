```python
from sklearn.preprocessing import StandardScaler
```

```python
Scaler=StandardScaler()
```

```python
# Standardize the numerical columns
scaler = StandardScaler()
df_scaled = df.copy()
numerical_columns = ['Academic Pressure', 'CGPA', 'Financial Stress', 'Work Pressure', 'Study Satisfaction']
df_scaled[numerical_columns] = scaler.fit_transform(df[numerical_columns])
```

```python
for column in numerical_columns:
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x=df_scaled[column], y=df['Depression'])
    plt.title(f'{column} vs Depression (Scaled)')
    plt.xlabel(column)
    plt.ylabel('Depression')
    plt.show()
```

```python
for column in numerical_columns:
    plt.figure(figsize=(8, 6))
    sns.regplot(x=df_scaled[column], y=df['Depression'], scatter_kws={'s': 20, 'alpha': 0.5})
    plt.title(f'{column} vs Depression with Regression Line')
    plt.xlabel(column)
    plt.ylabel('Depression')
    plt.show()
```

```python
categorical_columns = ['Gender', 'City', 'Profession', 'Degree', 'Have you ever had suicidal thoughts ?', 'Family History of Mental Illness']
```

```python
for column in categorical_columns:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=df[column], y=df['Depression'])
    plt.title(f'{column} vs Depression')
    plt.xlabel(column)
    plt.ylabel('Depression')
    plt.xticks(rotation=90)
    plt.show()
```

```python
sns.pairplot(df[['Academic Pressure', 'CGPA', 'Financial Stress', 'Depression']], hue='Depression', palette='coolwarm')
plt.title("Pairplot of Features vs Depression")
plt.show()
```
